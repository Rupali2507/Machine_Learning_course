# Data Visualization with Matplotlib

Matplotlib is a powerful Python library for creating static, animated, and interactive visualizations. It is widely used for plotting data and exploring datasets visually. In this notebook, we will demonstrate how to use Matplotlib to create various types of plots for data analysis.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]


plt.plot(x,y,color='red',linestyle=':',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## multiple plot 
## sample data

x=[1,2,3,4,5]
y1=[1,2,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
# row col
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,x,color='black')
plt.title("Plot 4")
```

```python
# bar plots

categories=['A','B','C','D','E']
values=[5,7,3,8,6]


plt.bar(categories,values,color="blue")
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar Plot')
plt.show()
```

## Histogram

A histogram is a graphical representation that organizes a group of data points into user-specified ranges. It is useful for visualizing the distribution of numerical data. In the next cell, we will create a histogram using the `values` list.

Histogram are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
data=[1,2,3,3,3,3,4,5,5,5,5,6,6,7,7,7,7,7,7]

plt.hist(data,bins=5,color='Orange',edgecolor='black')
```

```python
### create a scatter plot

x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color="blue",marker='x')
```

```python
# Pie charts

labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)


plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```
