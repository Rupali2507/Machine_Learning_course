# Seaborn Data Visualization

Seaborn is a powerful Python library for creating informative and attractive statistical graphics. It is built on top of matplotlib and integrates closely with pandas data structures. Seaborn makes it easy to generate complex visualizations with just a few lines of code.

**Key Features:**
- High-level interface for drawing attractive statistical graphics
- Built-in themes for improved aesthetics
- Functions for visualizing univariate and bivariate distributions
- Tools for fitting and visualizing linear regression models
- Support for categorical variables and multi-plot grids



Seaborn is ideal for quick exploration and visualization of data, making it a valuable tool in any data science workflow.

```python
import seaborn as sns
```

```python
#basic plotings

tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs tip")
plt.show()
```

```python
#line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of total_bill vs size")
```

```python
#categorical

sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#box plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## kde plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#pair plot

sns.pairplot(tips)
```

```python
#heatmap

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
